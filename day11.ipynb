```python
class person:
    def __init__(self,name,city):
        self.name=name
        self.city=city
    def get__info(self):
        print(f"{self.name} is the name and lives in {self.city} ")
```

```python
class student(person):
    def __init__(self,name,city,age,group):
        super().__init__(name,city)
        self.age=age
        self.group=group
    
    def get_details(self):
        print(f"the name of the student is{self.name},age is {self.age} and the group he took it {self.group}")
    
```

```python
s1=student("vijay","hyd","20","mpc")
s1.get_details()
```
