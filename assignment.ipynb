```python
# reverse a number 
num=int(input("give me another num:"))
rev=0
while num > 0:
    digit = num % 10
    rev = rev * 10 + digit
    num //= 10

print(rev)
```

```python
# minutes to seconds 
hr=int(input("give me time in hrs:"))
num=int(input("give me time in minutes: "))
hr_min=hr*60
num=num+hr_min
print(f"{num*60} is the time in seconds")
```
