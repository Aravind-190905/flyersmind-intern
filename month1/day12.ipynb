```python
class account:
    def __init__(self,name,accountno,accountbalence:int):
        self.name=name 
        self.accountno=accountno 
        self.accountbalence=accountbalence
        
    def get_balence(self):
        print(f"the balence in the account {self.accountno} is{self.accountbalence}")
        
    def deposit(self,amount):
        self.accountbalence=self.accountbalence+amount
        print(f"the amount {amount} is added")

    def withdraw(self,amount):
        if self.accountbalence-amount >0:
            self.accountbalence=self.accountbalence-amount
            print(f"the amount has been deduced")   
        else :
            print("insufficient funds ") 
    
```

```python
class savingaccount(account):
    def __init__(self, name, accountno, accountbalence,intrestrate):
        super().__init__(name, accountno, accountbalence)
        self.intrestrate=intrestrate
    def addinterest(self):
        self.accountbalence=self.accountbalence+(self.accountbalence*self.intrestrate)/100


        
```

```python
a1=account("aravind","1234",200)
a1.deposit(1000)
a1.get_balence()
```

```python
a2=savingaccount("aravind","1234",200,5)
a2.addinterest()
a2.get_balence()
```
