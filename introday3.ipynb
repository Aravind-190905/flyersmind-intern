```python
# make a prime or composite number detectot 
a=int(input("give me a number :"))
b=0
for i in range(2,a):
    if a % i ==0:
        print("its composite")
        break
    else:
        b=1

if b==1:
    print("its prime")
```

```python
# lets make a pattern that is a right triangle 
for i in range(0,3):
    for j in range(0,3):
        if j <=i:
            print("*",end=" ")

    print('\n')
    
```
