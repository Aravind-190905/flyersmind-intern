```python
marks=[]
n=int(input("give me the number of elements :"))
for i in range(0,n):
    marks.append(int(input("the marks are : ")))
```

```python

total=0
for i in marks:
    total=total+i
total
```

```python
avg=total/len(marks)
avg
```

```python
grades=[]
for i in marks:
    if i  > 80 :
        grade="A+"
    elif i > 60 and i <80 :
        grade="A"
    elif i >40 and i <60:
        grade="B+"
    elif i >30 and i <40:
        grade="pass"
    else:
        grade="dawg u got cooked☠️☠️"
    grades.append(grade)
grades
```

```python
# sort 
marks.sort()
marks
```

```python
#search
m=int(input("give me the number u wanna find :"))
for i in marks:
    if m == i:
        print(f"{m} is at index {marks.index(m)}")
```
